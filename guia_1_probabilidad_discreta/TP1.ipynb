Simular dos variables aleatorias $X_1$ y $X_2$ que siguen distribuciones de Poisson con parametros $\mu_1 = 1$ y $\mu_2 = 2$ respectivamente. Calcular la variable $Y = X_1 + X_2$. Repetir la simulaci칩n 1000 veces. Graficar la distribuci칩n de la variable $Y$ y comparar con una distribuci칩n de probabilidad adecuada.

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.special import factorial # from math import factorial
from numpy.random import poisson as np_poisson

def poisson(mu, k):
    return np.exp(mu) * np.power(mu, k)/factorial(k)
```

```python
N = 1000 # Numero de experimentps
mu1 = 1 # Valor medio X1
mu2 = 2 # Valor medio X2
```

```python
Y = [np_poisson(mu1) + np_poisson(mu2) for i in range(N)]
bins, high = np.unique(Y, return_counts=True)

teorical = poisson(mu1 + mu2, bins)
```

```python
plt.figure(figsize=(8, 5))

plt.ylabel("Frecuencia", size=18)
plt.xlabel("Y", size=18)
plt.title("Suma de poissons", fontsize=20)


plt.bar(bins, high/sum(high), alpha=0.7, color="darkcyan", label="Frecuencia observada")
plt.plot(bins, teorical/sum(teorical),  "--", color="k", alpha=0.8, label="Poisson(mu1 + mu2)" )

plt.legend(fontsize=14)
plt.show()
```
